# Aritmatika Dasar dengan Numpy

```python
import numpy as np

# Matriks arr
arr = np.array([[1, 2, 3], [5, 6, 7]], dtype=np.int64)

# Matriks arr_rdm (dengan elemen-elemen acak)
arr_rdm = np.random.randint(0, 100, size=(2, 3), dtype=np.int64)

# Operasi penjumlahan arr dan arr_rdm
addition_result = arr + arr_rdm

# Operasi pengurangan arr dan arr_rdm
subtraction_result = arr - arr_rdm

# Operasi perpangkatan dua arr_rdm
power_of_two_result = arr_rdm ** 2

# Operasi perkalian elemen arr dan arr_rdm
elementwise_multiplication_result = arr * arr_rdm

# Transpose matriks arr_rdm
transposed_arr_rdm = arr_rdm.T

# Operasi perkalian matriks arr dan arr_rdm (dot product)
matrix_multiplication_result = np.dot(arr, arr_rdm.T)  # Perbaikan di sini

# Mencetak hasil operasi
print("Hasil penjumlahan arr dan arr_rdm:")
print(addition_result)

print("Hasil pengurangan arr dan arr_rdm:")
print(subtraction_result)

print("Hasil perpangkatan dua arr_rdm:")
print(power_of_two_result)

print("Hasil perkalian elemen arr dan arr_rdm:")
print(elementwise_multiplication_result)

print("Transpose matriks arr_rdm:")
print(transposed_arr_rdm)

print("Hasil perkalian matriks arr dan arr_rdm (dot product):")
print(matrix_multiplication_result)
```

```python
import numpy as np

# Matriks arr
arr = np.array([[1, 2, 3], [5, 6, 7]], dtype=np.int64)

# Menghitung nilai rata-rata seluruh matriks arr
mean_arr = np.mean(arr)

# Menghitung nilai rata-rata per baris matriks arr
mean_per_row = np.mean(arr, axis=1)

# Menghitung total nilai seluruh elemen matriks arr
total_sum = np.sum(arr)

# Menghitung nilai maksimum per kolom matriks arr
max_per_column = np.max(arr, axis=0)

# Mencetak hasil
print("Nilai rata-rata seluruh matriks arr:", mean_arr)
print("Nilai rata-rata per baris matriks arr:", mean_per_row)
print("Total nilai seluruh elemen matriks arr:", total_sum)
print("Nilai maksimum per kolom matriks arr:", max_per_column)
```

# Fitur pada Numpy

```python
import numpy as np

# Membuat array arrg dengan nilai range 1-20
arrg = np.arange(1, 21)

# Mencetak array arrg
print("Array arrg:", arrg)

# Mencetak nilai arrg pada index 5
print("Nilai arrg index 5:", arrg[5])

# Mencetak nilai arrg untuk 3 elemen dari belakang
print("3 elemen dari belakang:", arrg[-3:])

# Mencetak nilai arrg dari indeks 3 sampai 6 (indeks 3, 4, 5, dan 6)
print("Nilai arrg dari indeks 3 sampai 6:", arrg[3:7])
```

Logika Program : 

```python
Program ini menggunakan library NumPy untuk membuat, mengakses, dan mencetak informasi dari sebuah array. Berikut adalah penjelasan logika dari program tersebut:

import numpy as np: Ini mengimpor library NumPy dan memberinya alias 'np' agar kita dapat mengakses fungsi-fungsi NumPy menggunakan 'np'.
arrg = np.arange(1, 21): Baris ini membuat array arrg yang berisi nilai dari 1 hingga 20. 
Fungsi np.arange(start, stop) digunakan untuk membuat array dengan nilai berurutan dari start (inklusif) hingga stop (eksklusif).

print("Array arrg:", arrg): Ini mencetak seluruh array arrg ke layar.
print("Nilai arrg index 5:", arrg[5]): Ini mencetak nilai yang terdapat pada index 5 dari array arrg. Ingatlah bahwa indeks dimulai dari 0, 
jadi nilai ini adalah elemen keenam dari array.
print("3 elemen dari belakang:", arrg[-3:]): Ini mencetak 3 elemen terakhir dari array arrg 
menggunakan indeks negatif. -1 adalah elemen terakhir, -2 adalah elemen kedua dari belakang, dan seterusnya.
print("Nilai arrg dari indeks 3 sampai 6:", arrg[3:7]): Ini mencetak nilai dalam array arrg mulai dari indeks 3 hingga sebelum indeks 7 (indeks 3, 4, 5, dan 6). 
Range ini dijelaskan dengan notasi arrg[start:stop], dimana elemen dengan indeks start akan disertakan, tetapi elemen dengan indeks stop akan dikecualikan.

Dengan demikian, program ini menggunakan NumPy untuk membuat, mengakses, dan mencetak informasi dari array arrg.
```

```python
import numpy as np

# Matriks arr
arr = np.array([[1, 2, 3], [5, 6, 7]], dtype=np.int64)

# Mencetak matriks arr
print("Matriks arr:")
print(arr)

# Slicing matriks arr untuk menghasilkan output pertama
sliced_arr1 = arr[:, 1:]

# Mencetak hasil slicing pertama
print("Sliced arr (output pertama):")
print(sliced_arr1)

# Slicing matriks arr untuk menghasilkan output kedua
sliced_arr2 = arr[:, 2:]

# Mencetak hasil slicing kedua
print("Sliced arr (output kedua):")
print(sliced_arr2)
```

Logika Program :

```python
Program di bawah ini menggunakan NumPy untuk memanipulasi matriks dan melakukan operasi slicing. Berikut adalah logika programnya:

Mengimpor library NumPy dengan alias np. Membuat matriks arr dengan data yang diberikan. Matriks ini memiliki dua baris dan tiga kolom dengan tipe data int64.
Mencetak matriks arr untuk menampilkan isi matriksnya. Ini mencetak seluruh matriks arr.

Melakukan slicing pada matriks arr untuk menghasilkan "output pertama" (sliced_arr1). 
Slicing dilakukan dengan mengambil semua baris (:) dan kolom kedua hingga terakhir (1:). 
Dengan demikian, hasil slicing pertama akan berisi kolom kedua dan ketiga dari matriks arr.

Mencetak hasil slicing pertama (sliced_arr1) untuk menampilkan output pertama. Ini akan mencetak kolom kedua dan ketiga dari matriks arr.
Melakukan slicing pada matriks arr untuk menghasilkan "output kedua" (sliced_arr2). 
Slicing dilakukan dengan mengambil semua baris (:) dan kolom ketiga hingga terakhir (2:). Dengan demikian, hasil slicing kedua akan berisi kolom ketiga dari matriks arr.
Mencetak hasil slicing kedua (sliced_arr2) untuk menampilkan output kedua. Ini akan mencetak kolom ketiga dari matriks arr.

Jadi, program ini mencetak matriks awal arr, kemudian melakukan dua operasi slicing berbeda untuk menghasilkan output pertama dan output kedua, 
dan akhirnya mencetak hasil dari kedua operasi slicing tersebut.
```

```python
import numpy as np

# Membuat matriks 3x4 dengan tipe data int64 dan nilai elemen acak
mtx_rdm = np.random.randint(0, 100, size=(3, 4), dtype=np.int64)

# Mencetak bentuk matriks mtx_rdm
print("Bentuk mtx_rdm sebelum perubahan:", mtx_rdm.shape)

# Mengubah bentuk matriks mtx_rdm menjadi satu dimensi
mtx_rdm_1d = mtx_rdm.flatten()

# Mencetak bentuk matriks mtx_rdm setelah menjadi satu dimensi
print("Bentuk mtx_rdm setelah menjadi satu dimensi:", mtx_rdm_1d.shape)

# Mengubah bentuk matriks mtx_rdm menjadi 6x2
mtx_rdm_6x2 = mtx_rdm.reshape(6, 2)

# Mencetak bentuk matriks mtx_rdm setelah menjadi 6x2
print("Bentuk mtx_rdm setelah menjadi 6x2:", mtx_rdm_6x2.shape)
```

Logika Program :

```python
Program di atas menggunakan library NumPy untuk membuat, mengubah bentuk, dan mencetak informasi matriks. 
Berikut adalah penjelasan logika dari program tersebut:

import numpy as np: Seperti sebelumnya, ini mengimpor library NumPy dan memberinya alias 'np' agar kita dapat mengakses fungsi-fungsi NumPy menggunakan 'np'.
mtx_rdm = np.random.randint(0, 100, size=(3, 4), dtype=np.int64): 
Ini membuat matriks mtx_rdm dengan ukuran 3x4 (3 baris dan 4 kolom) yang berisi angka-angka acak antara 0 dan 99. Tipe data elemen-elemen matriks ini diatur ke int64.

print("Bentuk mtx_rdm sebelum perubahan:", mtx_rdm.shape): Ini mencetak bentuk (jumlah baris dan kolom) dari matriks mtx_rdm sebelum mengalami perubahan. 
shape adalah atribut matriks yang mengembalikan tupel berisi jumlah baris dan kolom.

mtx_rdm_1d = mtx_rdm.flatten(): Matriks mtx_rdm diubah menjadi satu dimensi menggunakan metode flatten(). Hasilnya disimpan dalam variabel mtx_rdm_1d.

print("Bentuk mtx_rdm setelah menjadi satu dimensi:", mtx_rdm_1d.shape): Ini mencetak bentuk dari matriks mtx_rdm_1d setelah diubah menjadi satu dimensi. 
Sekarang, mtx_rdm_1d adalah array satu dimensi.

mtx_rdm_6x2 = mtx_rdm.reshape(6, 2): Matriks mtx_rdm diubah menjadi matriks 6x2 menggunakan metode reshape(). 
Ini berarti matriks asli dengan 3 baris dan 4 kolom diubah menjadi matriks dengan 6 baris dan 2 kolom.

print("Bentuk mtx_rdm setelah menjadi 6x2:", mtx_rdm_6x2.shape): Ini mencetak bentuk dari matriks mtx_rdm_6x2 setelah mengalami perubahan bentuk. 
Sekarang, matriks ini memiliki 6 baris dan 2 kolom.

Dengan demikian, program ini membuat, mengubah bentuk, dan mencetak informasi dari matriks menggunakan NumPy.
```
